```python
# Define the path to your txt file
file_path = "/content/drive/MyDrive/Class/Untitled folder/alphabet.txt"

data_with_labels = []
buffered_data = []
current_label = None

# Open the txt file and read its contents
with open(file_path, 'r') as file:
    lines = file.readlines()

    for line in lines:
        # Check if the line contains a letter
        if any(char.isalpha() for char in line):
            current_label = line.strip().split("\t")[-1]
            for buffered_line in buffered_data:
                parts = buffered_line.split("\t")
                if len(parts) >= 4:  # Ensure there are enough parts to extract x, y, z
                    x, y, z = parts[1], parts[2], parts[3]
                    data_with_labels.append((current_label, x, y, z))
            buffered_data = []  # Clear the buffer
        else:
            buffered_data.append(line)

print(data_with_labels)
```

```python
import pandas as pd

# Convert the list of tuples into a DataFrame
df = pd.DataFrame(data_with_labels, columns=['Label', 'X', 'Y', 'Z'])

df
```

```python
from sklearn.preprocessing import LabelEncoder

predict = ['Label']
x = np.array(df.drop(predict, axis=1))
y = np.array(df[predict])

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

print(y_encoded)
```

```python
from numpy import mean
from numpy import std
from numpy import absolute
from sklearn.datasets import make_regression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
from sklearn.neighbors import KNeighborsRegressor

# define base model
model_k = KNeighborsRegressor(n_neighbors=2, weights='distance')#,algorithm='brute'
x_train, x_test, y_train, y_test =  train_test_split(x, y_encoded, test_size=0.2, shuffle=True) #train ;test
model_k.fit(x_train,y_train)


predictions = model_k.predict(x_test)
y_train_pred = model_k.predict(x_train)

from sklearn.metrics import mean_squared_error, r2_score
print('R squared on training set:')
print(r2_score(y_train, y_train_pred))

cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate the model and collect the scores
n_scores = cross_val_score(model_k, x, y_encoded, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force the scores to be positive
n_scores = absolute(n_scores)
# summarize performance
print('MAE with Orignal data: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))

from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = mean_squared_error(y_test, predictions, squared=False)

print(f'mae with test data: {mae}')
print(f'mse: {mse}')
print(f'rmse: {rmse}')

acc =  model_k.score(x_test, y_test)
print(acc)
```

```python
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, np.round(predictions).astype(int))
print(conf_matrix, '\n')

# Normalize
conf_matrix_n = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]

# Plot
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(conf_matrix_n, annot=True, fmt='.2f', cmap='YlGnBu',
            xticklabels=encoder.classes_, yticklabels=encoder.classes_)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Test Data Confusion Matrix')
plt.xticks(rotation=45)
plt.show(block=False)

# Metrics
print('\nF1 Score: {}'.format(f1_score(y_test, np.round(predictions).astype(int), average='macro')))
print('Precision: {}'.format(precision_score(y_test, np.round(predictions).astype(int), average='macro')))
print('Recall: {}'.format(recall_score(y_test, np.round(predictions).astype(int), average='macro')))
print('Accuracy: {}'.format(accuracy_score(y_test, np.round(predictions).astype(int))))
```
